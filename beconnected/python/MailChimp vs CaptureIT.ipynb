# MailChimp vs Capture IT
## Objective
To find out MailChimp subscribers who are not yet on Capture IT

```python
import pandas as pd
df_mc = pd.read_csv('mailchimp.csv')
df_ci = pd.read_csv('captureit.csv')
```

```python
df_mc.head()
```

```python
df_ci.head()
```

Find the overlap, i.e. subscribers of MailChimp that have already registered on Capture IT:

```python
dfovrlp = pd.merge(df_ci,df_mc,how='inner',on=['First Name','Last Name'])
```

```python
dfovrlp
```

```python
dfovrlp[['First Name','Last Name', 'Email Address']]
```

Export this overlap, note that in MailChimp, to access the people who are not on Capture IT, must use a `NOT` clause somewhere.

```python

dfovrlp[['First Name','Last Name', 'Email Address']].to_csv("ci_on_mc.csv",index=False)
```
